## 선형 리스트 생성 프로그램

```python
class Node():
    def __init__(self):
        self.data=None
        self.link=None
```

```python
def printNodes(start):
    current=start
    if current==None:
        return
    print(current.data,end="")
    while current.link!=None:
        current=current.link
        print(current.data,end="")
    print()
```

```python
def makeSimpleLinkedList(nameheight):
    global memory,head,current,pre
    #printNodes(head)
    node=Node()
    node.data=nameheight
    memory.append(node)
    if head==None:
        head=node
        return
    if head.data[1]<nameheight[1]:
        node.link=head
        head=node
        return
    current=head
    while current.link!=None:
        pre=current
        current=current.link
        if current.data[1]<nameheight[1]:
            node.link=current
            pre.link=node
            return
    current.link=node
```

```python
def deleteNode(deleteData):
    global memory,head,current,pre
    cnt = 1 # 노드의 순서를 세기 위한 카운터
    
    if deleteData == 1:
        current=head
        head=head.link
        del(current)
        return
    current=head
    while current.link!=None:
        cnt += 1
        pre=current
        current=current.link
        if cnt==deleteData:
            pre.link=current.link
            del(current)
            return
    print("삭제할 데이터가 존재하지 않습니다.")
```

```python
state = 0
memory=[]
head,current,pre=None,None,None

while(state != 3):
    state = int(input("선택하세요(1: 추가, 2: 삭제, 3: 종료)-->"))
    if state == 1:
        data = eval(input("추가할 데이터-->")) # tuple input
        makeSimpleLinkedList(data)
        printNodes(head)
    if state == 2:
        data = int(input("삭제할 순서(-->"))
        deleteNode(data)
        printNodes(head)
```
